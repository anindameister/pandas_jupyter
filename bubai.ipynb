```python
import pandas as pd
```

```python
df = pd.read_csv('just image15python.csv')
```

```python
df
```

```python
# object_id_list = []
y_values = []
relationship = []

for _,row in df.iterrows():
#     object_id_list.append(row['object id'])
    y_values.append(row['Y-centre'])

for _,row in df.iterrows():
#     object_id_list.append(row['object id'])
    image_values.append(row['Image name'])


```

```python
def remove_value(current_element_index, current_list):
    if current_element_index==0:
        new_list = current_list[1:]
        return new_list
    new_list = current_list[:current_element_index-1]+current_list[current_element_index:]
    return new_list
```

```python
for index,current_element in enumerate(y_values):
    new_y_values = remove_value(index, y_values)
    for new_element in new_y_values:
        if current_element<new_element:
            relationship.append(f'object {y_values.index(current_element)} is at the top of object {y_values.index(new_element)}')
        if current_element>new_element:
            relationship.append(f'object {y_values.index(current_element)} is at the bottom of object {y_values.index(new_element)}')
```
